The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1  
 3: 1,3  
 6: 1,2,3,6  
10: 1,2,5,10  
15: 1,3,5,15  
21: 1,3,7,21  
28: 1,2,4,7,14,28  
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
import math
```

```python
#lembra o problema 3 onde usava a raiz do numero pra calcular o primo, mas aqui retornarÃ¡ uma lista dos primos
def fatorar(n):      
    vet = [] 
    for i in range(1,int(math.sqrt(n))+1): #mesmo racional problema 3
        if (n%i == 0):
            vet.append(int(n/i)) #aqui appendara o 'espelho'
            vet.append(i) #aqui appendara todos os primos
            #no fim, vetor esta appendando em pares
    if(math.sqrt(n) == int(math.sqrt(n))): #esse espelho duplica o resultado quando raiz quadrada = natural
        vet.pop()
    return vet
```

```python
i = 0
while True:
    i = i + 1
    n = int((i*(i+1)/2))
    vet = fatorar(n)
    #print(vet)
    if(len(vet) >= 500):
        print(n)
        break
```
